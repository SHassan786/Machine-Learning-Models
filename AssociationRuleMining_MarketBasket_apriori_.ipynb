```python
import string

transactionsList = []

for customer in range(0,7500):
    transaction = []
    for item in range(0,20):
        if str(data.values[customer,item]) != 'nan':
            transaction.append(str(data.values[customer,item]))
    transactionsList.append(transaction)
#print(transactionsList)
```

```python
from efficient_apriori import apriori
itemsets, rules = apriori(transactionsList, min_support=0.03, min_confidence=0.2,max_length=3)
print(len(rules))
```
